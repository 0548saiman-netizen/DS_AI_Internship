# Mini Project 1 – Exploratory Data Analysis
## Introduction
This dataset contains customer analytics data used to understand customer behavior and business performance.
Each row in the dataset represents a single customer record.  
The columns include various attributes related to customer demographics, purchasing behavior, and engagement metrics.
The purpose of this analysis is to explore the structure of the dataset, identify missing or inconsistent values, and uncover patterns that can provide insights into customer trends and business performance.

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

## Duplicate Removal
Duplicate rows were identified to prevent repeated customer records from skewing analysis results.
Any duplicate entries were removed to ensure data accuracy and reliability.
Maintaining unique customer records ensures that statistical summaries and visualizations reflect true patterns in the dataset.

```python
plt.figure()
plt.hist(df['Age'], bins=10)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

The annual income distribution appears moderately spread out.
Most customers fall within a mid-income range, with fewer customers in extremely high-income brackets.

```python
plt.figure()
plt.hist(df['YearsEmployed'], bins=10)
plt.title("Distribution of YearsEmployed")
plt.xlabel("YearsEmployed")
plt.ylabel("Frequency")
plt.show()
```

### Distribution of Years Employed
The histogram shows how the number of years employees have been employed is distributed.  
Most employees have been employed for **[insert observation, e.g., 1–10 years]**, with fewer employees in the higher experience ranges.  
This suggests that the majority of the workforce is relatively new, while a smaller portion are long-term employees.

```python
plt.figure()
df['Gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

The gender distribution shows whether the customer base is balanced or skewed.
If one gender dominates, it may indicate a targeted or niche market segment.

```python
plt.figure()
plt.scatter(df['YearsEmployed'], df['PurchaseFrequency'])
plt.title("YearsEmployed vs PurchaseFrequency")
plt.xlabel("YearsEmployed")
plt.ylabel("PurchaseFrequency")
plt.show()
```

The scatter plot shows how PurchaseFrequency relates to YearsEmployed.
If points are widely scattered, it suggests YearsEmployed alone does not determine PurchaseFrequency.

```python
plt.figure()
sns.boxplot(x='Gender', y='PurchaseFrequency', data=df)
plt.title("PurchaseFrequency by Gender")
plt.show()
```

The boxplot compares Purchase Frequency between genders.
Differences in median or spread may indicate varying purchasing tendencies across gender groups.

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
correlation_matrix = numeric_df.corr()
```

```python
correlation_matrix
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

The correlation heatmap helps identify relationships between multiple numerical variables simultaneously.
Strong positive correlations (values close to +1) indicate that two variables increase together.
Strong negative correlations (values close to -1) indicate an inverse relationship.
Weak correlations (values near 0) suggest little to no linear relationship between variables.
