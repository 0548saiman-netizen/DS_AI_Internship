# Task 1

Machine Learning is a way of programming where, instead of explicitly telling the computer all the rules to follow, we give it examples of data along with the correct answers and let it discover the patterns on its own. In traditional programming, humans carefully write rules that transform inputs into outputs. In Machine Learning, we reverse that process: we provide many input–output examples, and the system learns the underlying rules by identifying patterns in the data. This allows computers to solve problems that are too complex to define with fixed rules, such as recognizing images, understanding speech, or predicting trends.

# Supervised Learning

Definition:

Input (features)
Correct output (label/answer)
The model learns a mapping from inputs → outputs.

Common Tasks:
Classification (spam vs. not spam)
Regression (predicting price)

# Unsupervised Learning

Definition:
Unsupervised learning uses unlabeled data.
There are no correct answers provided — the model must find patterns, structure, or clusters on its own.

Common Tasks:
Clustering
Dimensionality reduction
Anomaly detection

# Reinforcement Learning (RL)

Definition:
Reinforcement learning involves an agent interacting with an environment, learning through:
Rewards (positive feedback)
Penalties (negative feedback)
There are no direct correct answers — only feedback signals.

Common Elements:
Agent
Environment
Actions
Rewards
Policy
